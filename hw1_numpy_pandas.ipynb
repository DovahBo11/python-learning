## Part 1: NumPy Basics (20 points)

### Question 1.1 (10 points)

Given the following array of weekly hours worked:

```python
hours = np.array([40, 52, 35, 45, 60, 38, 42, 55, 48, 37])
```

a) Calculate and print the mean, min, and max hours

b) Find all values greater than 45 (use boolean indexing)

c) Count how many people worked more than 45 hours

````python
import numpy as np

hours = np.array([40, 52, 35, 45, 60, 38, 42, 55, 48, 37])

# a) Calculate and print the mean, min, and max hours
MeanOfHours = np.mean(hours)
MinOfHours = np.min(hours)
MaxOfHours = np.max(hours)
print("A)\nMean hours:", MeanOfHours, "\nMin hours:", MinOfHours, "\nMax hours:", MaxOfHours, "\n")

# b) Find all values greater than 45 (use boolean indexing)
TargetHours = hours[hours > 45]
print("B)\nHours greater than 45:", TargetHours, "\n")

# c) Count how many people worked more than 45 hours
TargetHoursCount = np.size(TargetHours)
print("C)\nNumber of people who worked more than 45 hours:", TargetHoursCount)
````

### Question 3.1 (10 points)

a) How many rows and columns are in the dataset?

b) Which columns have missing values and how many?

c) What percentage of people earn >50K?

````python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
           'marital_status', 'occupation', 'relationship', 'race',
           'sex', 'capital_gain', 'capital_loss', 'hours_per_week',
           'native_country', 'income']

df = pd.read_csv(url, names=columns, na_values=' ?', skipinitialspace=True)

# a) How many rows and columns are in the dataset?
df_shape = df.shape
print("A)\nrows: ", df_shape[0], "\ncolumns: ", df_shape[1], "\n")

# b) Which columns have missing values and how many?
missing_values = df.isnull().sum()
print("B)\nColumns with missing values:\n", missing_values)


# c) What percentage of people earn >50K?
PercentOfPeople = df[df['income'] == '>50K'].shape[0] / df.shape[0] * 100
print("C)\npeople: ", PercentOfPeople)
````

### Question 3.2 (10 points)

a) What is the average age in the dataset?

b) What is the average hours_per_week?

c) What are the top 3 most common education levels?

````python
# a) What is the average age in the dataset?
Average_Age = df['age'].astype(float).mean()
print("A)\nAverage age:", Average_Age, "\n")

# b) What is the average hours_per_week?
Avarage_HrsPerWeek = df['hours_per_week'].astype(float).mean()
print("B)\nAverage hours per week:", Avarage_HrsPerWeek, "\n")

# c) What are the top 3 most common education levels?
TopEduLevels = df['education'].value_counts().head(3).index.tolist()
# '.value_counts().head(3)' gets the 3 most common values in the ['education'] column
# '.index()' only takes the value names and not their counts
# '.tolist()' converts the index to a list to avoid errors in printing
print("C)\nTop 3 most common education levels:\n", TopEduLevels)
````

### Question 3.3 (10 points)

a) Filter the dataset for people aged 30 or younger. How many are there?

b) Filter for people with income '>50K' AND hours_per_week > 40. How many?

c) What percentage of people work in the 'Private' sector?

````python
# a) Filter the dataset for people aged 30 or younger. How many are there?
df_Under30 = df[df['age'] <= 30]
# due to the fact 'age' is originally stored as an integer, it can be filtered directly
# since df_Under30 is a dataframe, we can use .shape[0] to get the number of rows hence the number of people
print("A)\nPeople aged 30 or younger:", df_Under30.shape[0], "\n")

# b) Filter for people with income '>50K' AND hours_per_week > 40. How many?
df_IncomeFiltered = df[(df['income'] == '>50K')]
df_HoursFiltered = df[df['hours_per_week'] > 40]
df_TargetFiltered = df_IncomeFiltered.index.intersection(df_HoursFiltered.index)

print("B)\nNumber of people with income '>50K' AND hours_per_week > 40:", df_TargetFiltered.shape[0], "\n")

# c) What percentage of people work in the 'Private' sector?
df_WorkersInPrivate = df[df['workclass'] == 'Private']
PercentInPrivate = round(df_WorkersInPrivate.shape[0] / df.shape[0] * 100, 2)
print("C)\nPercentage of people working in 'Private' sector:", PercentInPrivate, "%")
````
